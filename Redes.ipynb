```python
import networkx as nx   #para generar las redes Erdös-Renyi y Barabasi-Albert
import matplotlib.pyplot as plt  #para las gráficas
N=10000
kmed=6
```

```python
p_ER=kmed/(N-1)
red_ER = nx.erdos_renyi_graph(N, p_ER)   #red de Erdös-Renyi
nx.write_edgelist(red_ER, "red_ER.txt", data=False)
```

```python
m=kmed/2 #numero de enlaces nuevos que se forman es 3 porque queremos conectividad media 6
red_BA = nx.barabasi_albert_graph(N, m)
nx.write_edgelist(red_BA, "red_BA.txt", data=False)
```

```python
K_est=11
red_estrella=nx.star_graph(K_est-1)
nx.write_edgelist(red_estrella, "red_estrella.txt", data=False)
```

```python
#red BARABASI ALBERT PARA BANCO DE PECES
n_p=50
m=3 #numero de enlaces nuevos que se forman es 3 porque queremos conectividad media 6
red_BA_peces = nx.barabasi_albert_graph(n_p, m)
nx.write_edgelist(red_BA_peces, "red_BA_peces.txt", data=False)
```

```python
import numpy as np
import csv
from scipy.stats import binom

n = 9999
p = 6 / 9999
x = np.arange(0, 25)

# Distribución binomial real
binomial_pmf = binom.pmf(x, n, p)

datos = np.loadtxt('P(k)_ER_Bueno.txt', max_rows=25)
x = datos[:, 0]
y = datos[:, 1]

plt.plot(x, binomial_pmf, 'o-', label='Binomial (N=10000, p=6/9999)')
plt.plot(x, y, 'o-',label='ER simulación', color='orange')

plt.xlabel('k')
plt.ylabel('P(k)')
plt.title('Distribución P(k) para red ER con N=10000, p=6/9999')
plt.legend()
plt.grid(True)
plt.show()
```
