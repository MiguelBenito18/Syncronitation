```python
import networkx as nx   #para generar las redes Erdös-Renyi y Barabasi-Albert
import matplotlib.pyplot as plt  #para las gráficas
N=10000
kmed=6
```

```python
p_ER=kmed/(N-1)
red_ER = nx.erdos_renyi_graph(N, p_ER)   #red de Erdös-Renyi
nx.write_edgelist(red_ER, "red_ER.txt", data=False)
```

```python
m=3 #numero de enlaces nuevos que se forman es 3 porque queremos conectividad media 6
red_BA = nx.barabasi_albert_graph(N, m)
nx.write_edgelist(red_BA, "red_BA.txt", data=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson

n = 9999
p = 6 / 9999
x = np.arange(0, 25)

# Distribución binomial real
binomial_pmf = binom.pmf(x, n, p)

# Distribución de Poisson aproximada
#poisson_pmf = poisson.pmf(x, mu=6)

plt.plot(x, binomial_pmf, 'o-', label='Binomial(n=9999, p=6/9999)')
#plt.plot(x, poisson_pmf, 's--', label='Poisson(λ=6)', alpha=0.7)

plt.xlabel('x')
plt.ylabel('P(x)')
plt.title('Distribución Binomial')
plt.legend()
plt.grid(True)
plt.show()
```
